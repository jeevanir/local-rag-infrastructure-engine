```python
import os
from langchain_community.document_loaders import DirectoryLoader, TextLoader

def ingest_knowledge_base(directory_path: str):
    """
    Ingests all source text files from a directory into standard Document objects.
    """
    if not os.path.exists(directory_path):
        raise FileNotFoundError(f"Source folder '{directory_path}' not found.")

    print(f"[*] Initializing data ingestion pipeline for: {directory_path}")

    # Set up the loader to recursively search for text documents
    loader = DirectoryLoader(
        directory_path,
        glob="**/*.txt",
        loader_cls=TextLoader
    )

    raw_documents = loader.load()
    print(f"[вњ“] Ingestion Complete. Extracted {len(raw_documents)} raw document(s).")
    return raw_documents

# To test this, create a local folder named 'source_knowledge', drop a text file in, and run:
# raw_docs = ingest_knowledge_base("./source_knowledge")
```

```python
import os
from langchain_community.document_loaders import DirectoryLoader, TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

# === SYSTEM PATH CONFIGURATION ===
DIRECTORY_PATH = "./source_knowledge"

def execute_sprint_1_pipeline(directory_path: str):
    # 1. Environment Health Check / Auto-mocking data
    if not os.path.exists(directory_path):
        print(f"[*] Path '{directory_path}' not found. Creating it and seeding a mock data asset...")
        os.makedirs(directory_path)
        with open(os.path.join(directory_path, "sample_kb.txt"), "w") as f:
            f.write(
                "LangChain is a framework designed to simplify the creation of applications using large language models (LLMs).\n\n"
                "Retrieval-Augmented Generation (RAG) optimizes LLM output by referencing an authoritative knowledge base outside data sources before generating a response.\n\n"
                "Sprints help development teams manage complex engineering milestones with structural gravity."
            )

    # 2. RAG-6 Execution: Data Ingestion Pipeline
    print(f"\n[*] [RAG-6] Initializing data ingestion from: {directory_path}")
    loader = DirectoryLoader(
        directory_path,
        glob="**/*.txt",
        loader_cls=TextLoader
    )

    raw_documents = loader.load()
    print(f"[вњ“] [RAG-6] Success! Extracted {len(raw_documents)} raw document object(s).")

    # 3. RAG-7 Execution: Text Splitting Strategy
    print(f"\n[*] [RAG-7] Initializing text splitting strategy...")
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=100,      # Small chunk size for tracking exact splits clearly
        chunk_overlap=20,     # Standard overlapping sliding window
        separators=["\n\n", "\n", " ", ""]
    )

    final_chunks = text_splitter.split_documents(raw_documents)
    print(f"[вњ“] [RAG-7] Success! Segmented data into {len(final_chunks)} optimized text chunks.")

    # Let's inspect the system output
    print("\n--- System Inspection: First 2 Processed Chunks ---")
    for i, chunk in enumerate(final_chunks[:2]):
        print(f"Chunk {i+1} Metadata: {chunk.metadata}")
        print(f"Chunk {i+1} Content: '{chunk.page_content}'\n")

    return final_chunks

# Execute the complete operational pipeline
processed_chunks = execute_sprint_1_pipeline(DIRECTORY_PATH)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

def execute_standalone_rag7_splitting(documents, chunk_size=150, chunk_overlap=30):
    """
    RAG-7: Structural Text Splitting Engine
    Segments extracted document streams into mathematically optimized context chunks.
    """
    print(f"[*] [RAG-7] Launching text splitting strategy...")
    print(f"[*] Configuration Matrix -> Chunk Size: {chunk_size} chars | Overlap: {chunk_overlap} chars")

    # 1. Initialize the recursive strategy layer
    # Separators prioritize paragraph breaks, falling back to sentences, then spaces.
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        length_function=len,
        separators=["\n\n", "\n", " ", ""]
    )

    # 2. Execute segmentation
    split_chunks = text_splitter.split_documents(documents)

    print(f"[вњ“] [RAG-7] Segmentation Complete! Generated {len(split_chunks)} semantic chunks.")

    # 3. Structural Validation Panel
    print("\n" + "="*50)
    print("         SYSTEM INSPECTION: CHUNK ARCHITECTURE         ")
    print("="*50)
    for index, chunk in enumerate(split_chunks):
        print(f"\n[вЂў] CHUNK {index + 1} | Character Length: {len(chunk.page_content)}")
        print(f"    Source File: {chunk.metadata.get('source', 'Unknown')}")
        print(f"    Content Data:\n    \"{chunk.page_content}\"")
        print("-" * 50)

    return split_chunks

# --- Standalone Execution Path ---
# Assuming 'raw_documents' was populated by your RAG-6 DirectoryLoader:
# final_vector_ready_chunks = execute_standalone_rag7_splitting(raw_documents)
```
